## 1. Introducción
Mediante este ejercicio vamos a iniciarnos en la sintaxis de R y en la visualización de datos.

El procedimiento será el siguiente: primero cargaremos los datos desde un fichero _csv_, a continuación realizaremos una serie de operaciones para conocer mejor los datos que trataremos y finalmente los mostraremos gráficamente.

## 2. Formato csv
Un fichero _csv_ (_comma separated values_ o 'valores separados por coma') se trata de un formato de almacenamiento de datos en un fichero de texto plano. Cada línea del fichero representa un registro de información y cada registro está formado por uno o varios campos, separados por comas.

Un ejemplo de este formato es el siguiente:
    
    "Alex","Jorda","Cómics"
    "Jorge","Melguizo","Rugby"
    
No es un formato estandarizado y por eso no es extraño encontrar diferentes delimitadores como por ejemplo el carácter ';' (punto y coma). Además es común (y útil) que incluyan una cabezera describiendo la información del fichero, como es el caso del _csv_ con el que trabajaremos.

Para modificar un fichero con esta extensión __no__ se recomienda usar el programa _Excel_ puesto que este puede añadir carácteres no visibles y alterar los datos sin que nos demos cuenta.

### 2.1. Datos
El conjunto de datos que usaremos en este ejercicio de inicio es muy conocido y usado a mode de ejemplo/iniciación en temas de _machine learning_: el [_iris_ dataset](https://es.wikipedia.org/wiki/Iris_flor_conjunto_de_datos).

Este dataset recopila 150 muestras de tres especies distintas de Iris (un genero de plantas). Las medidas representan el largo y ancho de los sépalos y los pétalos de las distintas muestras, que hay 50 de cada especie.

___

## 3. Lectura del fichero
Como se ha comentado, el primer paso es la lectura del fichero _csv_. Usaremos la función _read.csv_ para ello, de la siguiente manera:

```r
data <- read.csv(file="iris.csv", sep=",", head=TRUE)
```

> Como iris es un dataset muy conocido, este viene incluido por R directamente.
  Otra manera de obtenerlo habría sido mediante:
  
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;dataset(iris)
        
> Donde tendríamos la información en la variable _iris_.

Ahora mismo tenemos toda la información volcada en la variable _data_. Para comprovar la correcta recuperación de la información podemos usar la función _head_, con la que veremos los primeros datos del _data frame_.

```r
head(data)
data
```

Como tenemos la información en la variable _data_, podemos hacer una pruebas de acceso a su información mediante índices.

```r
data[3,]    #Acceso a la tercera muestra
```

```r
data[,2]          #Acceso al segundo atributo (sepal.width)
identical(data[,2], data$sepal.width)    #Otra manera de acceder a un atributo es mediante su nombre
```

```r
data[5:10, c(2,4)]    #Acceso a los atributos 2 y 4 del rango de muestras de la 5 a la 10
```

## 4. Estadística
Ahora que ya tenemos los datos y hemos echado un vistazo a su estructuración, vamos a hacer otra aproximación obteniendo su media aritmética, su mediana, su variancia y su desviación estándar.

### 4.1. Primer contacto
En primera instancia, para obtener un vistazo general a los datos, es usual emplear la función _summary_ para una primera toma de contacto con el dataset.

```r
summary(data)
```

De esta manera podemos ver los atributos del conjunto y varios datos sobre ellos, como por ejemplo los mínimos, máximos y promedios.

### 4.2. Media aritmética o promedio
La función para obtener la media de un conjunto de valores es la función _mean_ (promedio), la cual recibe un vector y devuelve su promedio. 

A continuación, obtenemos el valor promedio del atributo sepal.length:

```r
mean(data$sepal.length)
```

Como nosotros conocemos la información de la que disponemos, sabemos que la distribución de las muestras es la siguiente:

    [1..50] -> Iris setosa
    [51..100] -> Iris versicolor
    [101..150] -> Iris virginica
   
podemos, por ejemplo, obtener el promedio de anchura del pétalo de la variedad versicolor:

```r
mean(data[1:50,]$petal.width)
mean(data[51:100,]$petal.width)
mean(data[101:150,]$petal.width)
```

### 4.3. Mediana 
La mediana es otro dato estadístico interesante. Para obtenerla se usa la función _median_ (mediana), que funciona exactamente igual que _mean_: recibe un vector y devuelve un valor.

```r
median(data$sepal.length)
```

### 4.4 Varianza y desviación estándar
Otras magnitudes interesantes son la varianza y la desviación típica.

Estas mesuras se obtienen mediante las funciones _var_ (variance) y _sd_ (standard deviation).

```r
var(data[101:150,]$petal.width)
sd(data[101:150,]$petal.width)
```

## 5. Representación gráfica
Una parte de gran importancia en el tratamiento de datos es claramente su visualización gráfica. De esta manera podemos ver con más facilidad ciertas correlaciones entre diferentes datos.

A continuación, vamos a mostrar las relaciones de todos los atributos entre si.

```r
plot(data[,])
```

Mediante esta representación podriamos ver si hay algun atributo que nos ayude a discernir entre las tres variedades.

En este caso, observamos que mediante los atributos petal.length y petal.width podemos diferenciar la primera clase (Setosa) de las otras dos.

Claramente, a la función plot, si en vez proveerla con _data_ entera le proporcionamos solo dos atributos, se nos mostrará un único gráfico como el siguiente:

```r
plot(data[,3:4])
```

> En el caso anterior también podríamos haber usado:

> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;plot(data\$petal.width, data\$petal.length)

> O incluso para mostrar la relación entre los atributos 3 y 4, pero únicamente de la variedad Setosa:

> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;plot(data[1:50,3:4]) 

La función _plot_ es muy interesante de investigar, ya que tiene diversos parámetros que pueden modificar considereblemente el aspecto del resultado, por ejemplo el parámetro _type_.

Para finalizar, también podemos generar histogramas usando la función _hist_.

> Si el parámetro _type_ de la función _plot_ tiene el valor de "h", también se pueden generar histogramas.

```r
hist(data$petal.length[1:50], xlab="Petal length", ylab="Frequency", main="Petal length in Iris Setosa", col="blue")
```
