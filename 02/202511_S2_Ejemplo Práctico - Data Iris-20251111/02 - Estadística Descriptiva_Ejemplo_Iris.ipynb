## Desviación estándar y varianza

La desviación estándar y la varianza se calculan con las funciones sd()y var():


```r
sd(dat$Sepal.Length) # Desviación estándar (standard deviation)
```

```r
var(dat$Sepal.Length) # Varianza
```

La desviación estándar y la varianza son diferentes si la calculamos para una muestra o una población. En R, la desviación estándar y la varianza se calculan como si los datos representaran una muestra (por lo que el denominador es <i>n - 1</i>).

Demostración: http://www3.uah.es/marcos_marva/files_bioest/2016-17/varianza_sesgada.html

### Consejo: 

Para calcular la desviación estándar (o varianza) de varias variables al mismo tiempo, utilizar lapply() indicando las estadísticas adecuadas como segundo argumento:


```r
lapply(dat[, 1:4], sd)

#La función lapply, aplica una función dada a varios conjuntos de datos
#Parámetros (lapply(param1, param2)):
#param1: son los conjuntos de datos pasados como columnas de un dataset
#param2: función a aplicar a los conjuntos de datos
```

El comando dat[, 1:4] selecciona las columnas de la 1 a la 4 ya que la quinta variable es una variable cualitativa y la desviación estándar no se puede calcular en ese tipo de variable. Este comando funciona de la siguiente manera:

nombre_dataset[número_de_fila, número_de_columna] -> para acceder a un elemento en concreto de la tabla
nombre_dataset[rango_de_filas,] -> para obtener los datos de las filas introducidas
nombre_dataset[,rango_de_columnas] -> para obtener los datos de las columnas introducidas

Y a partir de aquí se pueden realizar combinaciones de los parámetros para obtener la información que nos interese.

## Tabla de contingencia

En estadística las tablas de contingencia se emplean para registrar y analizar la asociación entre dos o más variables, habitualmente de naturaleza cualitativa (nominales u ordinales).

La función table() también se puede utilizar en dos variables cualitativas para crear una tabla de contingencia. El conjunto de datos iris tiene solo una variable cualitativa, por lo que creamos una nueva variable cualitativa solo para este ejemplo.

Creamos la variable <b>size</b> que corresponde a <b>small</b> si la longitud del pétalo es menor que la mediana de todas las flores, en <b>big</b> caso contrario:


```r
dat$size <- ifelse(dat$Sepal.Length < median(dat$Sepal.Length),
  "small", "big"
)

#La función ifelse() comprueba la condición definida en el primer parámetro y, si se cumple, devuelve el valor 
#   definido en el segundo parámetro y si no, devuelve el valor definido en el tercer parámetro
#
#En estas líneas de código se está creando una nueva variable dentro del dataset 'dat', llamada 'size', los valores
#   de la cual vendrán dados por el resultado de la función ifelse(). Esta función se aplica a todos los regitros de
#   la variable seleccionada
```

Aquí hay un resumen de las ocurrencias por tamaño:

```r
table(dat$size)
```

Ahora creamos una tabla de contingencia de las dos variables Species y size con la función table():

```r
table(dat$Species, dat$size)
```

o con la función xtabs():

```r
xtabs(~ dat$Species + dat$size)
```

La tabla de contingencia da el número de casos en cada subgrupo.
Por ejemplo, solo hay una flor de setosa grande, mientras que hay 49 flores de setosa pequeñas en el conjunto de datos.

Para ir más allá, podemos ver en la tabla que las flores de setosa parecen tener un tamaño más grande que las flores de virginica.
Para comprobar si el tamaño se asocia significativamente con la especie, podríamos realizar una prueba de independencia de Chi-cuadrado ya que ambas variables son variables categóricas (próxima sesión ;-) ).

Tener en cuenta que Species están en filas y size en columna porque especificamos Species y luego size en table().

En lugar de tener las frecuencias (es decir, el número de casos), también se pueden tener las frecuencias relativas (es decir, proporciones) en cada subgrupo agregando la función table() dentro de la función prop.table():


```r
prop.table(table(dat$Species, dat$size))
```

Tener en cuenta que también puede calcular los porcentajes por fila o por columna agregando un segundo argumento a la función prop.table(): 1 para fila o 2 para columna:

```r
# Porcentajes por fila:
round(prop.table(table(dat$Species, dat$size), 1), 2) 

# La función round(), redondea los números según el número de decimales que se le especifica, en este caso 2.
```

```r
# Porcentajes por columna:
round(prop.table(table(dat$Species, dat$size), 2), 2) 
```

## Histograma

Un histograma da una idea sobre la distribución de una variable <b>cuantitativa</b>.

La idea es dividir el rango de valores en intervalos y contar cuántas observaciones caen en cada intervalo.

```r
hist(dat$Sepal.Length,
       col=c("#7facff"),
       border="#668957",
       main= "Histograma de la variable 'Sepal.Length'",
       xlab="Longitud del Sépalo",
       ylab="Frecuencia")
```

Si agregamos los argumentos "breaks =" dentro de la función hist() si queremos cambiar el número de bins. (Más info. en: https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/hist)

Una regla empírica (conocida como ley de Sturges) [2] es que el número de bins debe ser el valor redondeado de la raíz cuadrada del número de observaciones. El conjunto de datos incluye 150 observaciones, por lo que en este caso el número de bins se puede establecer en 12.


En {ggplot2}:

```r
ggplot(dat) +
  aes(x = Sepal.Length) +
  geom_histogram(
      fill = "#7facff", 
      color = "white")+ 
  labs(title="Distribución de la variable 'Sepal.Length'" #título del gráfico
      ,subtitle = "núm.bins = default"
      ,x = "Longitud del Sépalo")  #nombre variable eje x
```

De forma predeterminada, el número de contenedores es 30.

Se puede cambiar este valor con, geom_histogram(bins = 12) por ejemplo.

```r
ggplot(dat) +
  aes(x = Sepal.Length) +
  geom_histogram(
      bins=12,  #establecemos número de bins
      fill = "#7facff", 
      color = "white")+ 
  labs(title="Distribución de la variable 'Sepal.Length'" #título del gráfico
      ,subtitle = "núm.bins = default"
      ,x = "Longitud del Sépalo")  #nombre variable eje x
```

## Boxplot

<u>Los diagramas de caja son realmente útiles en la estadística descriptiva</u> y, a menudo, se subutilizan (principalmente porque el público no los comprende bien).

Un BoxPlot representa gráficamente la distribución de una variable cuantitativa al mostrar visualmente cinco resúmenes de ubicaciones comunes (mínimo, mediana, primer / tercer cuartiles y máximo) y cualquier observación que se clasificó como un valor atípico sospechoso utilizando el criterio de rango intercuartílico (IQR).


Antes de dibujar una gráfica de caja de nuestros datos, vea a continuación un gráfico que explica la información presente en una gráfica de caja:



```r
boxplot(dat$Sepal.Length,
       col=c("#93c47d"),
       border="#668957",
       main= "Boxplot de la variable 'Sepal.Length'",
       ylab="Longitud del Sépalo")
```

Los diagramas de caja son aún más informativos cuando se presentan uno al lado del otro para comparar y contrastar distribuciones de dos o más grupos. Por ejemplo, comparamos la longitud del sépalo en las diferentes especies:


```r
boxplot(dat$Sepal.Length ~ dat$Species,
       col=c("#93c47d"),
       border="#668957",
       main= "Boxplot de la variable 'Sepal.Length'",
       ylab="Longitud del Sépalo")
```

En {ggplot2}:


```r
ggplot(dat) +
  aes(x = Species, y = Sepal.Length) +
  geom_boxplot(
      fill = "#93c47d", 
      color = "white")+ 
  labs(
      title="Boxplot de la variable 'Sepal.Length'" #título del gráfico
      ,y = "Longitud del Sépalo")
```

## Scatterplot

Los diagramas de dispersión permiten comprobar si existe un vínculo potencial entre dos variables cuantitativas. Por esta razón, los diagramas de dispersión se utilizan a menudo <b>para visualizar una posible correlación entre dos variables</b>.

Por ejemplo, al dibujar un diagrama de dispersión de la longitud del sépalo y la longitud del pétalo:

```r
plot(dat$Sepal.Length, dat$Petal.Length,
       pch = 21,
       bg="#F79068",
       main= "Scatterplot de la variable 'Sepal.Length'",
       xlab= "Longitud del Sépalo",
       ylab= "Longitud del Pétalo")
```

Parece haber una asociación positiva entre las dos variables.

En {ggplot2}:

```r
ggplot(dat) +
  aes(x = Sepal.Length, y = Petal.Length) +
  geom_point(
      color = "#F79068")+ 
  labs(
      title="Scatterplot de la variable 'Sepal.Length'" #título del gráfico
      ,x = "Longitud del Sépalo"
      ,y = "Longitud del Pétalo")
```

Al igual que los diagramas de caja, los diagramas de dispersión son aún más informativos al diferenciar los puntos según un factor, en este caso la especie:

```r
ggplot(dat) +
  aes(x = Sepal.Length, y = Petal.Length, colour = Species) +
  geom_point()+ 
  labs(
      title="Scatterplot de la variable 'Sepal.Length'" #título del gráfico
      ,x = "Longitud del Sépalo"
      ,y = "Longitud del Pétalo") +
  scale_color_hue()
```

## Gráfico QQ

<b>Para una sola variable</b>

Para verificar el supuesto de normalidad de una variable (normalidad significa que los datos siguen una distribución normal, también conocida como distribución gaussiana), usualmente usamos histogramas y / o gráficos QQ.

Los histogramas se han presentado anteriormente, así que aquí se muestra cómo dibujar un gráfico QQ:

```r
# Grafica los puntos:
qqnorm(dat$Sepal.Length)
# Grafica la línea de referencia:
qqline(dat$Sepal.Length)
```

O un gráfico QQ con bandas de confianza con la función qqPlot() del paquete{car}:

```r
install.packages("car")
```

```r
library(car) # package must be installed first
```

```r
qqPlot(dat$Sepal.Length)
```

Si los puntos están cerca de la línea de referencia (a veces denominada línea de Henry) y dentro de las bandas de confianza, se puede considerar que se cumple el supuesto de normalidad.

Cuanto mayor sea la desviación entre los puntos y la línea de referencia y cuanto más se encuentren fuera de las bandas de confianza, es menos probable que se cumpla la condición de normalidad. 

La variable Sepal.Length no parece seguir una distribución normal porque varios puntos se encuentran fuera de las bandas de confianza.

Cuando se enfrenta a una distribución no normal, el primer paso suele ser aplicar la transformación logarítmica a los datos y volver a verificar si los datos transformados logarítmicamente se distribuyen normalmente.
La aplicación de la transformación logarítmica se puede hacer con la función log().
